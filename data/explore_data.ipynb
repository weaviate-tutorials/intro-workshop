```python
import pandas as pd
from pathlib import Path
import json

data_file = Path("tmdb_data_1950_2024.json")
data = json.loads(data_file.read_text())
```

```python
year_dfs = list()
for k, v in data.items():
    tmp_df = pd.DataFrame(v)
    tmp_df["year"] = k
    year_dfs.append(tmp_df)
```

```python
df = pd.concat(year_dfs)
df = df[df["vote_count"] > 2000]
```

```python
assert len(df[df["adult"] == True]) == 0
```

```python
df.drop(columns=["adult"], inplace=True)
```
